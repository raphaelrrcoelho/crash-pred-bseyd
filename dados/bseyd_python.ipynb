```python
import pandas as pd
import numpy as np
import math
from scipy.stats import norm
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
dados = pd.read_excel('./BSEYD_EUA_dados.xlsx', engine='openpyxl', index_col='Dates')
```

```python
dados
```

```python
dados['spx_bseyd'] = dados['gt10_govt'] - dados['spx_earnings']
```

```python
dados
```

```python
dados['mean_movel'] = dados['spx_bseyd'].rolling(252).mean()
```

```python
dados['sd_movel'] = dados['spx_bseyd'].rolling(252).std()
```

```python
dados['z_score'] = (dados['spx_bseyd'] - dados['mean_movel'])/dados['sd_movel']
```

```python
dados['cl'] = dados['spx_bseyd'] + dados['z_score']*np.sqrt((dados['sd_movel']**2)/252)


"""
mean_movel - (norm_movel*(sqrt(sd_movel**2/252)))

"""
#dados['cl'] = dados['spx_bseyd'] + dados['z_score']*np.sqrt((dados['sd_movel']**2)/252)

dados
```

```python
dados

```

```python
for index, row in dados.iterrows():
    if row['z_score'] < row['cl']:
        dados.loc[index,'signal_bseyd'] = 1
    else:
        dados.loc[index,'signal_bseyd'] = 0
```

```python
"""
BSEYD_usa = 
  BSEYD_usa %>%
  mutate(signal_bseyd_teste22 = spx_bseyd - ci) %>%
  mutate(signal_bseyd_ss = ifelse(signal_bseyd_teste>0, 1,0))
"""
dados['signal_bseyd_teste'] = dados['spx_bseyd'] - dados['cl']
```

```python
for index, row in dados.iterrows():
    if row['signal_bseyd_teste'] > 0:
        dados.loc[index,'signal_bseydss'] = 1
    else:
        dados.loc[index,'signal_bseydss'] = 0

```

```python
"""
s1_teste = rollsum(BSEYD_usa$signal_bseyd_ss, 31, align='right', fill=NA)
BSEYD_usa = cbind(BSEYD_usa,s1_teste)

"""

dados['s1_teste'] = dados['signal_bseydss'].rolling(30).sum()
dados
```

```python
"""
 BSEYD_usa =
   BSEYD_usa %>%
   mutate(S = (ifelse((s1_teste==1) & (signal_bseyd_teste==1) ,1, 0)))

"""
for index, row in dados.iterrows():
    if (row['s1_teste']==1) & (row['signal_bseydss'] ==1):
        dados.loc[index,'s'] = 1
    else:
        dados.loc[index,'s'] = 0
```

```python
dados = dados.to_excel('dados.xlsx', index=True)
```
